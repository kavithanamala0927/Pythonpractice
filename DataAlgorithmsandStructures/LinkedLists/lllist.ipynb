```python
class Node:
    def __init__(self,data):
        self.data = data
        self.nextNode = None
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insertatHead(self,data):
        newNode = Node(data)
        newNode.nextNode = self.head
        self.head = newNode        
    
    def insertatTail(self,data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            return self.head    
        current = self.head
        while current.nextNode != None:
            current = current.nextNode
        current.nextNode = newNode
        
    def insert_After_Node(self, node, data):
        current = self.head
        #print ("Current Node",current.nextNode.data)
        while current != None:
            #print ("data",current.data)
            if current.data == node:
                break 
            current = current.nextNode
            #print ("Current next",current.data)
        if current is None:
            print ("Item not in the list")
        else:
            #print ("else",current.data)
            newNode = Node(data)
            newNode.nextNode = current.nextNode
            #print ("newnode",newNode.nextNode)
            current.nextNode = newNode
    
    def insert_Before_Node(self,node,data):
        if self.head is None:
            print ("List is empty")
            return
        if node == self.head.data:
            newNode = Node(data)
            newNode.nextNode = self.head
            self.head = newNode
            return
        current = self.head
        #print (current.nextNode.data)
        while current.nextNode != None:
            if node == current.nextNode.data:
                break
            current = current.nextNode
        if current.nextNode is None:
            print ("Item not in list")
            return
        else:
            newNode = Node(data)
            newNode.nextNode = current.nextNode
            current.nextNode = newNode
    
    def insert_At_Index(self, index, data):
        if index == 1:
            newNode = Node(data)
            newNode.nextNode = self.head 
            self.head = newNode
        
        i = 1
        current = self.head
        while i < index-1 and current != None:
            current = current.nextNode
            i =i+1
        if current is None:
            print ("Index out of bound")
        else:
            newNode = Node(data)
            newNode.nextNode = current.nextNode
            current.nextNode = newNode
            
    def traverse(self,head):
        print("Elements in the linked list:::")
        if head is None:
            print ("LinkedList is empty.")
            return
        else:
            current = head
            while current != None:
                print (current.data,end="->")
                current = current.nextNode
    def reverse_LinkedList(self,head):
        print ("Reverse LinkedList:::")
        previous = None
        current = head
        #print ("current",current.data)
        while current != None:
            next_node = current.nextNode
            current.nextNode = previous
            previous = current
            #print ("previous",previous.data)
            current = next_node
        head = previous
        return head
    def print_reverse(self):
        print ("Printing in reverse order::")
        if self.head is None:
            return

        items = []
        current = self.head
        while current != None:
            items.append(current.data)
            current = current.nextNode
        while items:
            print (items.pop())
        
    def deleteAtIndex(self,index):
        print ("Deleting Element:::")
        if self.head is None:
            return 
        if index == 0:
            self.head = self.head.nextNode
            return
        current = self.head
        i = 1
        while current.nextNode != None and i <= index :
            if i == index:
                break
            i = i+1
            current = current.nextNode
        if current.nextNode is None:
            print ("Item not found in the list")
        else:
            current.nextNode = current.nextNode.nextNode
            
    def getNodeatIndex(self,index):
        count = 0
        result = self.head
        current = self.head
        while (count <= index-1):
            current = current.nextNode
            count += 1
        while current.nextNode != None:
            current = current.nextNode
            result = result.nextNode
        print ("Element",result.data)
            
    def mergeLists(list1,list2):
        list1 = list1.head
        list2 = list2.head
        if list1 is None:
            return list2
        if list2 is None:
            return list1
        if list1 == None and list2 == None:
            return None
        if list1 and list2:
            if list1.data <= list2.data:
                sorting = list1
                list1.nextNode = sorting.nextNode
            else:
                sorting = list2
                list2.nextNode = sorting.nextNode
        result = sorting
        while list1 and list2:
            if list1.data <= list2.data:
                sorting.nextNode = list1
                sorting = list1
                list1 = sorting.nextNode
            else:
                sorting.nextNode = list2
                sorting = list2
                list2 = sorting.nextNode
        return result
    def get_nthnode_from_end(self,node):
        p = self.head
        f = self.head
        count = 1
        while count<=(node-1):
            f = f.nextNode
            count += 1
        while f.nextNode != None:
            p = p.nextNode
            f = f.nextNode
        return p.data
    def delete_node(self,deleteData):
        p = self.head.nextNode
        q = self.head
        if q.data == deleteData:
            self.head = p
        else:
            while p.data != deleteData:
                p = p.nextNode
                q = q.nextNode
            if p.nextNode == None:
                q.nextNode = None
            else:
                q.nextNode = p.nextNode
        return self.head
    def removeNthFromEnd(self,head,n):
        slow = fast = head
        count =0
        while fast and count < n:
            fast = fast.nextNode
            count = count+1
        if fast == None:
            temp = head
            if head is not None:
                head = head.nextNode
            temp = None
            return head
        prev_of_slow = None
        while slow and fast:
            prev_of_slow = slow
            slow = slow.nextNode
            fast = fast.nextNode
        prev_of_slow.nextNode = slow.nextNode
        slow = None
        return head
    def check_Palindrome(self,head):
        slow = fast = head
        start_second = None
        while fast and fast.nextNode:
            fast = fast.nextNode.nextNode
            if fast == None:
                start_second = slow.nextNode
                break
            if fast.nextNode == None:
                start_second = slow.nextNode.nextNode
                break     
            slow = slow.nextNode
        slow.nextNode = None
        #Reverse the second list
        previous = None
        current = start_second
        while current!=None:
            next_node = current.nextNode
            current.nextNode = previous
            previous = current
            current = next_node
        start_second = previous
        #Compare both the lists
        while head and start_second:
            if head.data != start_second.data:
                return 0
            head = head.nextNode 
            start_second = start_second.nextNode
        return 1

    def rotateRight(self,head,n):
        size = 1
        current = head
        while current.nextNode:
            current = current.nextNode
            size+=1
        n = n%size
        fast = slow =head
        count = 0
        if n==0:
            return head
        while fast.nextNode and count<n:
            fast = fast.nextNode
            count+=1
        while fast.nextNode:
            fast = fast.nextNode
            slow = slow.nextNode
        new_head = slow.nextNode
        slow.nextNode = None
        temp = new_head
        while temp.nextNode:
            temp = temp.nextNode
        temp.nextNode = head
        return new_head
    def getSize(self,head):
        current = head
        size = 1
        while current.nextNode:
            current = current.nextNode
            size += 1
        return size
            
    def addTwoLinkedLists(self,list1,list2):
        carry = 0
        temp = None
        prev = None
        print ("head",self.head)
        while list1 is not None or list2 is not None:
            if list1 is None:
                list1_data  = 0
            else:
                list1_data = list1.data
            if list2 is None:
                list2_data = 0
            else:
                list2_data = list2.data
            list_sum = carry+list1_data+list2_data
            if list_sum >= 10:
                carry = 1
            else:
                carry = 0
            if list_sum < 10:
                list_sum = list_sum
            else:
                list_sum = list_sum % 10
            temp = Node(list_sum)
            if self.head is None:
                self.head = temp
            else:
                prev.nextNode = temp
            prev = temp
            if list1.nextNode:
                list1 = list1.nextNode
            if list2.nextNode:
                list2 = list2.nextNode
        if carry > 0:
            temp.nextNode = Node(carry)
        return temp
                
    def addTwoLists(self, list1, list2): 
        sum_list = list1.data + list2.data
        carry = int(sum_list/10)
        result = Node(sum_list%10)
        p3 = result
        p1 = list1.nextNode 
        p2 = list2.nextNode
        while p1 is not None or p2 is not None:
            sum_list = carry+(p1.data if p1 else 0)+(p2.data if p2 else 0)
            carry = int(sum_list/10)
            p3.nextNode = Node(sum_list%10)
            p3 = p3.nextNode
            p1 = p1.nextNode if p1 else None
            p2 = p2.nextNode if p2 else None
        if carry > 0 :
            p3.nextNode = Node(carry)
        return result
    
    
    def reverseSubList(self, head, m, n):
        if not head:
            return None

        current, prev = head, None
        while m > 1:
            prev = current
            current = current.nextNode
            m, n = m - 1, n - 1

        tail, con = current, prev

        while n:
            third = current.nextNode
            current.nextNode = prev
            prev = current
            current = third
            n -= 1

        if con:
            con.nextNode = prev
        else:
            head = prev
        tail.nextNode = current
        return head
    def reorderList(self,head):
        current = head
        deque = []
        
        while current:
            deque.append(current.data)
            current = current.nextNode
        i= 0
        temp = head
        while deque:
            if i%2 == 0:
                temp.data = deque.pop(0)
            else:
                temp.data = deque.pop()
            i += 1
            temp = temp.nextNode
        return head
    def swapPairwise(self,head):
        current = head
        if not current or not current.nextNode :
            return head
        while current and current.nextNode:
            current.data,current.nextNode.data = current.nextNode.data, current.data
            
            current = current.nextNode.nextNode
        return head
    def swapPairs(self, A):
        if A and A.nextNode :
            head = A.nextNode
        else:
            head = A
        
        while A and A.nextNode:
            C = A.nextNode.nextNode 
            #print ("C",C.data)# 3rd Node
            D = C.nextNode if C and C.nextNode else C  
            #print ("D",D.data)
            A.nextNode.nextNode, A.nextNode, A = A, D, C
            #print ("A.nextnode",A.nextNode.nextNode.data)
          
        return head
    def reverseGroupbyK(self,head,k):
        next_Node = None
        prev = None
        current = head
        count = 0
        while current and count < k:
            next_Node = current.nextNode
            current.nextNode = prev
            prev= current
            current = next_Node
            count += 1
        if next_Node:
            head.nextNode = self.reverseGroupbyK(next_Node,k)
        return prev
    def detectCycle(self, head):
        current = head
        loop = head
        cycle = False
        while current and loop and loop.nextNode:
            current = current.nextNode
            loop = loop.next.nextNode
            if current == loop:
                cycle = True
                break
        if cycle:
            current = head
            while(current != loop):
                current = current.nextNode
                loop = loop.nextNode

            return current
        return None
        
        
    def partitionList(self,head,x):
        small= small_head = Node(0)
        big = big_head = Node(0)
        #current = head
        
        while head:
            print (head.data)
            if head.data < x:
                print ("IF")
                small.nextNode = head
                small = small.nextNode
                #current = current.nextNode
            else:
                print ("ELSE")
                big.nextNode = head
                big = big.nextNode
            head = head.nextNode
        big.nextNode = None
        small.nextNode = big_head.nextNode
        return small_head.nextNode
    
    
    def insertionSortLinkedList(self,head):
        sorted_list = None
        current = head
        while current:
            next_node = current.nextNode
            if sorted_list == None or sorted_list.data >= current.data:
                current.nextNode = sorted_list
                sorted_list = current
            else:
                pointer = sorted_list 
                while pointer.nextNode and pointer.nextNode.data < current.data:
                    pointer = pointer.nextNode
                current.nextNode = pointer.nextNode
                pointer.nextNode = current 

            current = next_node
        return sorted_list     
    def mergeSort(self,head):
        #Timecomplexity O(nlogn)
        if head == None or head.nextNode == None: 
            return head
  
        # get the middle of the list  
        middle = self.getMiddle(head)
        
        nexttomiddle = middle.nextNode
  
        # set the next of middle node to None 
        middle.nextNode = None

        # Apply mergeSort on left list  
        left = self.mergeSort(head) 
        # Apply mergeSort on right list 
        right = self.mergeSort(nexttomiddle) 
        return right
         
        # Merge the left and right lists  
        sortedlist = self.sortedMerge(left, right) 
        return sortedlist 
    
    def getMiddle(self, head): 
        if (head == None): 
            return head 
  
        slow = head 
        fast = head 
  
        while (fast.nextNode != None and 
               fast.nextNode.nextNode != None): 
            slow = slow.nextNode
            fast = fast.nextNode.nextNode
              
        return slow 
    def sortedMerge(self, a, b): 
        result = None
          
        # Base cases 
        if a == None: 
            return b 
        if b == None: 
            return a 
              
        # pick either a or b and recur.. 
        if a.data <= b.data: 
            result = a 
            result.nextNode = self.sortedMerge(a.nextNode, b) 
        else: 
            result = b 
            print ("A and B.nextNode",a.data, b.nextNode.data)
            result.nextNode = self.sortedMerge(a, b.nextNode) 
        return result
        

```
